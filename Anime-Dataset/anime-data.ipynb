```python
import requests
import pandas as pd
```

#### Creating Anime Dataset using Jikan API (Unofficial API of MyAnimeList)

```python
anime_list = []
page = 1

while True:
    res = requests.get(f"https://api.jikan.moe/v4/anime?page={page}")
    data = res.json()

    for anime in data["data"]:
        anime_list.append({
            "mal_id": anime["mal_id"],   
            "title": anime["title"],
            "type": anime.get("type"),   
            "episodes": anime.get("episodes"),
            "score": anime.get("score"),
            "popularity":anime.get("popularity"),
            "rank":anime.get("rank"),
            "status": anime.get("status"),
            "genre": anime.get("genres")[0]["name"]
        })
    if page!=15:
        page += 1
    else:
        break
```

```python
df = pd.DataFrame(anime_list)
df.to_csv("anime_dataset.csv", index=False)
print("Dataset saved successfully!")
```
