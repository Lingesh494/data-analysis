```python
from faker import Faker
import random
import pandas as pd

f=Faker()
```

### We are going to create a dataset containing fake data of students

```python
name=[]
email=[]
phone=[]
gender_r=["Male","Female"]
uni_r=["MIT","Harvard","Cambridge","Yale"]
gender=[]
uni=[]
math_marks=[]
eng_marks=[]
sci_marks=[]
math_grade=[]
eng_grade=[]
sci_grade=[]
```

```python
for i in range(300):
    g=random.choice(gender_r)
    if g=="Male":
        gender.append(g)
        first=f.first_name_male()
        last=f.last_name_male()
        n=first+" "+last
        name.append(n)
        em=f"{first.lower()}.{last.lower()}"+"@gmail.com"
        email.append(em)
        uni.append(random.choice(uni_r))
        phone.append(f.phone_number())
        math_m=f.random_int(min=0,max=100)
        eng_m=f.random_int(min=0,max=100)
        sci_m=f.random_int(min=0,max=100)
        math_marks.append(math_m)
        eng_marks.append(eng_m)
        sci_marks.append(sci_m)
        if math_m >=90:
            math_grade.append("A")
        elif math_m >=80 and math_m < 90:
            math_grade.append("B")
        elif math_m >=70 and math_m <80:
            math_grade.append("C")
        elif math_m <70:
            math_grade.append("D")
            
        if eng_m >=90:
            eng_grade.append("A")
        elif eng_m >=80 and eng_m < 90:
            eng_grade.append("B")
        elif eng_m >=70 and eng_m <80:
            eng_grade.append("C")
        elif eng_m <70:
            eng_grade.append("D")              
        
        if sci_m >=90:
            sci_grade.append("A")
        elif sci_m >=80 and sci_m < 90:
            sci_grade.append("B")
        elif sci_m >=70 and sci_m <80:
            sci_grade.append("C")
        elif sci_m <70:
            sci_grade.append("D")
            
    else:
        gender.append(g)
        first=f.first_name_female()
        last=f.last_name_female()
        n=first+" "+last
        name.append(n)
        em=f"{first.lower()}.{last.lower()}"+"@gmail.com"
        email.append(em)
        uni.append(random.choice(uni_r))
        phone.append(f.phone_number())
        math_m=f.random_int(min=0,max=100)
        eng_m=f.random_int(min=0,max=100)
        sci_m=f.random_int(min=0,max=100)
        math_marks.append(math_m)
        eng_marks.append(eng_m)
        sci_marks.append(sci_m)
        if math_m >=90:
            math_grade.append("A")
        elif math_m >=80 and math_m < 90:
            math_grade.append("B")
        elif math_m >=70 and math_m <80:
            math_grade.append("C")
        elif math_m <70:
            math_grade.append("D")
            
        if eng_m >=90:
            eng_grade.append("A")
        elif eng_m >=80 and eng_m < 90:
            eng_grade.append("B")
        elif eng_m >=70 and eng_m <80:
            eng_grade.append("C")
        elif eng_m <70:
            eng_grade.append("D")              
        
        if sci_m >=90:
            sci_grade.append("A")
        elif sci_m >=80 and sci_m < 90:
            sci_grade.append("B")
        elif sci_m >=70 and sci_m <80:
            sci_grade.append("C")
        elif sci_m <70:
            sci_grade.append("D")
```

```python
data={
    "Name":name,
    "Gender":gender,
    "Email":email,
    "Phone":phone,
    "University":uni,
    "Math_Marks":math_marks,
    "English_Marks":eng_marks,
    "Science_Marks":sci_marks,
    "Math_Grade":math_grade,
    "English_Grade":eng_grade,
    "Science_Grade":sci_grade
}
df=pd.DataFrame(data)
df.to_csv("students_dummy_data.csv",index=False)
```
